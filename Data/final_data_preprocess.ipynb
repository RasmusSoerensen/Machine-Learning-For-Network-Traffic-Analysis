```python
#Do the preprocessing of the data, except OneHotEncoding and Standardization
output_file = 'C:/Users/Rasmu/OneDrive/DTU/Kunstig intelligens og Data/7. semestre/Bachelor/Machine-Learning-For-Network-Traffic-Analysis/Data/combined_data/final_data.csv'
first = True
for i in all_file_path:
  for chunk in pd.read_csv(i, chunksize=10000000, delimiter='\t', comment='#', header=None): #Change the chunksize to a smaller number if you run out of memory or incrase it if you have enough memory
    chunk.columns = ["ts", "uid", "id.orig_h", "id.orig_p", "id.resp_h", "id.resp_p", "proto", "service", "duration", "orig_bytes", "resp_bytes", "conn_state", "local_orig", "local_resp","missed_bytes", "history", "orig_pkts", "orig_ip_bytes", "resp_pkts", "resp_ip_bytes", "tunnel_parents_label_detailed-label"]
    chunk[['tunnel_parents', 'label', 'detailed_label']] = chunk['tunnel_parents_label_detailed-label'].str.split(expand=True)
    chunk.drop(['tunnel_parents_label_detailed-label'], axis=1, inplace=True)
    chunk['label'] = chunk['label'].replace({'Benign': 0, 'benign': 0, 'malicious': 1, 'Malicious': 1})
    chunk.drop(["ts","uid","service","local_orig", "local_resp","missed_bytes","history","tunnel_parents", "id.orig_h", "id.resp_h", "id.orig_p", "id.resp_p","orig_ip_bytes","resp_ip_bytes"], axis=1, inplace=True)
    duplicates = chunk.duplicated()
    chunk = chunk.drop_duplicates()
    chunk.replace({'-': np.nan, '(empty)': np.nan}, inplace=True)
    #If you feel for a better mapping of the labels, you can change the mapping here
    detailed_label_mapping = {np.nan: 0,'C&C': 1,'C&C-Mirai': 1,'DDoS': 2,'PartOfAHorizontalPortScan': 3,'PartOfAHorizontalPortScan-Attack': 3,'C&C-PartOfAHorizontalPortScan': 3,
        'C&C-FileDownload': 4, 'FileDownload': 4,'C&C-HeartBeat-FileDownload': 4,'Okiru': 5,'Okiru-Attack': 5,'C&C-Torii': 6,'C&C-HeartBeat': 7,'Attack': 8}
    chunk['detailed_label'] = chunk['detailed_label'].replace(detailed_label_mapping)
    chunk.dropna(inplace=True)
    #Standardize the data
    scaler = StandardScaler()
    chunk[['duration', 'orig_bytes', 'resp_bytes', 'orig_pkts', 'resp_pkts']] = scaler.fit_transform(chunk[['duration', 'orig_bytes', 'resp_bytes', 'orig_pkts', 'resp_pkts']])
    #Ensure that only the first time the data is written to the file, the header is written
    if first:
      chunk.to_csv(output_file, index=False)
      first = False
      del chunk
      gc.collect() #Clean up the memory
    else:
      chunk.to_csv(output_file, mode='a', header=False, index=False)
      del chunk
      gc.collect()

#One-hot encoding
data = pd.read_csv(output_file)
features = ['proto', 'conn_state']
encoding = OneHotEncoder(sparse_output=False)
encoding.fit(data[features])
onehotlabels = encoding.transform(data[features])
onehotlabels_df = pd.DataFrame(onehotlabels, columns=encoding.get_feature_names_out(features))
data = data.drop(features, axis=1)
data = pd.concat([data, onehotlabels_df], axis=1)
data['detailed_label'].value_counts()
```

```python
#Sample the datasets to get a balanced dataset and an imbalanced dataset, removing the labels 4 and 6 since there are too few of them
data = data[data.detailed_label != 4]
data = data[data.detailed_label != 6]
b_equall = data[data['label'] == 0].sample(n=77898, random_state=1) #77898 is the half number of benign samples (see output from previous cell)
m_equall = data[data['label'] == 1].sample(n=77898, random_state=1)
data_balanced = pd.concat([b_equall, m_equall])
#Ensure no duplicates in the data
data-b_equall
data-m_equall
b_imbalanced = data[data['label'] == 0].sample(n=77897, random_state=1)
m_imbalanced = data[data['label'] == 1].sample(n=7789, random_state=1)
data_imbalanced = pd.concat([b_imbalanced, m_imbalanced])

#Write the data to a file
data_balanced.to_csv('C:/Users/Rasmu/OneDrive/DTU/Kunstig intelligens og Data/7. semestre/Bachelor/Machine-Learning-For-Network-Traffic-Analysis/Data/combined_data/final_data_balanced.csv', index=False)
data_imbalanced.to_csv('C:/Users/Rasmu/OneDrive/DTU/Kunstig intelligens og Data/7. semestre/Bachelor/Machine-Learning-For-Network-Traffic-Analysis/Data/combined_data/final_data_imbalanced.csv', index=False)
```

```python
#Forward Feature Selection
data = pd.read_csv('C:/Users/Rasmu/OneDrive/DTU/Kunstig intelligens og Data/7. semestre/Bachelor/Machine-Learning-For-Network-Traffic-Analysis/Data/combined_data/final_data_balanced.csv')
#Ensure datasplit
X = data.drop(['label', 'detailed_label'], axis=1)
y = data['label']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

#Train the model
clf = DecisionTreeClassifier(random_state=1)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
importances = clf.feature_importances_ #Get the feature importances
indices = np.argsort(importances)[::-1]

#Plot the feature importances
plt.figure(figsize=(10, 6))
plt.bar(range(X_train.shape[1]), importances[indices])
plt.xticks(range(X_train.shape[1]), [X.columns[i] for i in indices], rotation=90)
plt.ylabel('Feature Importance')
plt.show()
```
